1. What is the difference between a function and a method in Python?

  - In Python, functions and methods are both blocks of reusable code that perform specific tasks, but they differ primarily in how they are associated with objects.

ðŸ”¹ Function
A function is defined using the def keyword.

It is not associated with any object unless explicitly passed.

Can be called independently.
Example:

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Output: Hello, Alice!
```

ðŸ”¹ Method
A method is a function that is associated with an object, typically within a class.

Called on an object (instance) and often uses the self parameter to access instance data.

There are instance methods, class methods, and static methods in classes.

Example:

```python
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Bob"))  # Output: Hello, Bob!
```

5. What are iterators in Python and how do they differ from iterables?

  - In Python:

An iterable is any object you can loop over (e.g., list, string, tuple).

An iterator is an object that remembers its state and yields items one at a time using the __next__() method.

You get an iterator from an iterable by using the iter() function.
Example:

```python
nums = [1, 2, 3]         # This is an iterable
it = iter(nums)          # This is an iterator

print(next(it))  # Output: 1
print(next(it))  # Output: 2
print(next(it))  # Output: 3
```

6. Explain the concept of generators in Python and how they are defined.

  - Generators in Python are special functions that yield values one at a time using the yield keyword instead of return. They produce items lazily, which means they generate values on the fly and are memory-efficient.

Example:

```python
def count_up_to(n):
    for i in range(1, n + 1):
        yield i

gen = count_up_to(3)
print(next(gen))  # Output: 1
```

7. What are the advantages of using generators over regular functions?

- Generators offer several key advantages over regular functions, especially when dealing with large datasets or streams of data:

âœ… 1. Memory Efficiency
Generators yield one item at a time, without storing the entire sequence in memory.

Ideal for working with large data or infinite sequences.

âœ… 2. Lazy Evaluation
Values are produced on demand, reducing computation and memory usage until needed.

âœ… 3. Improved Performance
Faster startup time compared to building large data structures at once.

Can be paused and resumed, making them efficient for pipelines.

âœ… 4. Clean Syntax
Generators use yield, which makes the code more readable than managing iteration state manually.

âœ… 5. Infinite Sequences
You can easily define infinite sequences (e.g., an infinite number generator) without running out of memory.

ðŸ”¹ Example Scenario:
A regular function that returns a list of 1 million numbers consumes a lot of memory. A generator yields one number at a time, using almost no memory regardless of the range size.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

   - The functions map(), reduce(), and filter() in Python are used for functional-style programming. Here's a clear comparison:

Function	Purpose	Returns	Example Use Case
map()	Applies a function to each item in an iterable	Map object (iterable)	Transforming values (e.g., squaring numbers)
filter()	Filters items based on a condition	Filter object (iterable)	Selecting values (e.g., even numbers)
reduce()	Applies a function cumulatively to items	Single value	Reducing to a result (e.g., summing all numbers)

Example of each:

```python
from functools import reduce

nums = [1, 2, 3, 4, 5]

# map: square each number
squares = list(map(lambda x: x**2, nums))  # [1, 4, 9, 16, 25]

# filter: get even numbers
evens = list(filter(lambda x: x % 2 == 0, nums))  # [2, 4]

# reduce: sum all numbers
total = reduce(lambda x, y: x + y, nums)  # 15
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
nums = [10, 15, 20, 25, 30]
print(sum_of_even_numbers(nums))  # Output: 60
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage:
for power in powers_of_two(5):
    print(power)
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # strip() removes any trailing newlines or spaces
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 37, 100]

# Convert to Fahrenheit using map and a lambda function
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))

# Print the result
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Input string
input_string = "Hello, how are you doing today?"

# Define vowels
vowels = "aeiouAEIOU"

# Use filter() to remove vowels
filtered_string = ''.join(filter(lambda ch: ch not in vowels, input_string))

# Print the result
print("String without vowels:", filtered_string)
```
